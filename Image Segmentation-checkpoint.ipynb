```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
```

```python
img = cv2.imread('elephant.jpg')
plt.imshow(img)
plt.show()
```

```python
#converting image bgr to rgb
img = cv2.imread('elephant.jpg')
img = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
plt.imshow(img)
original_img = img
plt.show()
```

```python
print(img.shape)
```

```python
#flatten each channel of the image in linear array
all_pixel=img.reshape((330*500,3))
```

```python
print(all_pixel.shape)
```

```python
from sklearn.cluster import KMeans
```

```python
dominant_colors = 4
km = KMeans(n_clusters=dominant_colors)
km.fit(all_pixel)
```

```python
centers = km.cluster_centers_
```

```python
centers = np.array(centers, dtype='uint8')
```

```python
print(centers)
```

```python
#plot all colors are these values
```

```python
i=1

colors = []
for each_col in centers:
    plt.subplot(1,4,i)
    plt.axis("off")
    i+=1
    
    colors.append(each_col)
    
    #color swatch
    
    a = np.zeros((100,100,3), dtype='uint8')
    a[:, :, :] = each_col
    plt.imshow(a)
plt.show()
```

```python
#Ssegmenting our original image
```

```python
new_img = np.zeros((330*500,3), dtype='uint8')
print(new_img.shape)
```

```python
a=km.labels_
print(len(a))
colors
```

```python
for ix in range(new_img.shape[0]):
    new_img[ix] = colors[km.labels_[ix]]

new_img = new_img.reshape((original_img.shape))
```
